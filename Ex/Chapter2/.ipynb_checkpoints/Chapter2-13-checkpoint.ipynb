## 多項式回帰モデル

```python
import numpy as np
from sklearn import linear_model
import sklearn.metrics as sm
import matplotlib.pyplot as plt

%matplotlib inline

RANGE=5

# numpy.random.rand(): 0.0以上、1.0未満の一様分布の乱数を生成する
# rand(100)は乱数を100個生成するということ
# ここでRANGEは5を指定しているので、単純にMAX5までの乱数となる
x = np.random.rand(100) * RANGE

# xをもとにしたsin(x)に乱数（-0.3〜+0.3の範囲）を足している
y = np.sin(x) + np.random.rand(100) * 0.3
print(y)
# xは1行100列のベクトルとみなせるので、100行1列のベクトルを作成する
# reshapeの第1引数を-1とすると元のベクトルから列数を元にして行数を推定してくれる
x = x.reshape(-1, 1)
print(x)
# 線形回帰モデルを作成してx,yを元に訓練する
regressor = linear_model.LinearRegression()
regressor.fit(x, y)

#np.linspace(0, RANGE, 100)は0〜5まで等間隔の100個の値を生成する
x_test = np.linspace(0, RANGE, 100).reshape(-1, 1)

# 等間隔のx_testを元に線形回帰させる。線形なので直線になる
y_pred = regressor.predict(x_test)

# 出力を予測する
plt.scatter(x, y, color='green')
plt.plot(x_test, y_pred, color='black')

# グラフを表示する
plt.show()
```

### 多項式で近似して回帰させる

```python
from sklearn.preprocessing import PolynomialFeatures

# 多項式の乗数の部分をdegreeで表すdegree=10はx^10までの多項式で表すこと
# degree大きすぎると適合しすぎてしまう（過学習）
polynomial = PolynomialFeatures(degree=10)

# xを元に訓練とxを多次元（ここでは10次元）の値に調整したものをx_x_transformedとする
x_transformed = polynomial.fit_transform(x)

# 回帰モデルを作成して訓練する。多項式の場合、入力値xを多次元に調整してからyを算出する形になる
poly_linear_regressor = linear_model.LinearRegression()
poly_linear_regressor.fit(x_transformed, y)

# 検証データx_testからx_test_tranformedを10次元に調整したものを作成する
x_test_tranformed = polynomial.transform(x_test)

# 出力を予測する
y_pred = poly_linear_regressor.predict(x_test_tranformed)

# グラフを書く
plt.scatter(x, y, color='green')
plt.plot(x_test, y_pred, color='black')

# グラフを表示する
plt.show()
```
