### 混同行列

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# 正しいラベルセットと予測されたラベルセットの例を設定する
true_labels = [2, 0, 0, 2, 4, 4, 1, 0, 3, 3, 3] 
pred_labels = [2, 1, 0, 2, 4, 3, 1, 0, 1, 3, 3]

# ラベル値を作成
labels = [0, 1, 2, 3, 4]

# セットした真偽値を元に混同行列を作成する
confusion_mat = confusion_matrix(true_labels, pred_labels, labels=labels)
print(confusion_mat)

# 分かりやすくするため、pandasのデータフレームを使って、ラベル分けされた行列表示を作成する
cm_labeled = pd.DataFrame(confusion_mat, columns=labels, index=labels)
cm_labeled
```

```python
# 行列を渡すとその行列の要素を可視化してくれる
plt.imshow(confusion_mat, interpolation='nearest', cmap=plt.cm.gray)

# グラフのタイトル設定
plt.title('Confusion Matrix')

# カラーバーつける
plt.colorbar()

# ticksはラベルの数
ticks = np.arange(5)
plt.xticks(ticks, ticks)
plt.yticks(ticks, ticks)

# x軸、y軸の示すタイトル
plt.ylabel('True labels')
plt.xlabel('Predicted labels')
plt.show()
```

```python
targets = ['Class-0', 'Class-1', 'Class-2', 'Class-3', 'Class-4']
print(classification_report(true_labels, pred_labels, target_names=targets))
```
